```python
import math

radio1 = 3
radio2 = 8
radio3 = 10 

cincunferencia1 = 2 * math.pi * radio1
cincunferencia2 = 2 * math.pi * radio2
cincunferencia3 = 2 * math.pi * radio3

print ("la cincunferencia para un radio de", radio1, "es", cincunferencia1)
print ("la cincunferencia para un radio de", radio2, "es", cincunferencia2)
print ("la cincunferencia para un radio de", radio3, "es", cincunferencia3)

```
